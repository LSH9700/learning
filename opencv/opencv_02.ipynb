```python
# 비디오 회색으로바꾸고 옆으로붙임
import cv2
capture = cv2.VideoCapture('lookbook.mp4')
if capture.isOpened() ==False:
    print("동영상을 열수 없음")
    exit(1)

while cv2.waitKey(33) < 0:
    ret, img_src = capture.read()
    img_src = cv2.pyrDown(img_src)
    img_gray = cv2.cvtColor(img_src, cv2.COLOR_BGR2GRAY)
    img_gray = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR)
    img_dst = cv2.hconcat([img_src, img_gray])
    #동영상 끝까지 가면 처음으로 다시
    if capture.get(cv2.CAP_PROP_POS_FRAMES) == capture.get(cv2.CAP_PROP_FRAME_COUNT):
        capture.set(cv2.CAP_PROP_POS_FRAMES,0)
    cv2.imshow('dst',img_dst)
capture.release()
cv2.destroyAllWindows()
```

```python

# 영상 대칭
import cv2
img_src = cv2.imread('love.png',cv2.IMREAD_COLOR)
height, width = img_src.shape[:2]
#대칭 option = 0:상하  양수:좌우 음수:상하좌우
img_lr = cv2.flip(img_src,1)
img_ud = cv2.flip(img_src,0)
img_lr_ud = cv2.flip(img_src,-1)
img_dst = cv2.hconcat([img_src,img_lr])
img_tmp = cv2.hconcat([img_ud,img_lr_ud])
img_dst = cv2.vconcat([img_dst,img_tmp])
cv2.imshow('dst',img_dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#흑백영상 대칭
# 비디오 회색으로바꾸고 옆으로붙임
import cv2
capture = cv2.VideoCapture('lookbook.mp4')
if capture.isOpened() ==False:
    print("동영상을 열수 없음")
    exit(1)

while cv2.waitKey(33) < 0:
    ret, img_src = capture.read()
    img_src = cv2.pyrDown(img_src)
    img_gray = cv2.cvtColor(img_src, cv2.COLOR_BGR2GRAY)
    img_lr = cv2.flip(img_gray, 1)
    img_ud = cv2.flip(img_gray, 0)
    img_lr_ud = cv2.flip(img_gray, -1)
    img_dst = cv2.hconcat([img_gray, img_lr])
    img_tmp = cv2.hconcat([img_ud, img_lr_ud])
    img_dst = cv2.vconcat([img_dst, img_tmp])

    #동영상 끝까지 가면 처음으로 다시
    if capture.get(cv2.CAP_PROP_POS_FRAMES) == capture.get(cv2.CAP_PROP_FRAME_COUNT):
        capture.set(cv2.CAP_PROP_POS_FRAMES,0)
    cv2.imshow('dst',img_dst)
capture.release()
cv2.destroyAllWindows()
```

```python
# 이미지 회전
import cv2
img_src = cv2.imread('love.png',cv2.IMREAD_COLOR)
height, width = img_src.shape[:2]
center =(width//2, height//2)
angle = 90
matrix = cv2.getRotationMatrix2D(center,angle,1)
img_dst = cv2.warpAffine(img_src,matrix,(width,height))

cv2.imshow('dst',img_dst)
cv2.waitKey()
cv2.destroyWindow()
```

```python
# 이미지크기를 절반으로 줄이기
import numpy as np
import cv2

img_src = cv2.imread('lion.jpg',cv2.IMREAD_COLOR)
height, width = img_src.shape[:2]
img_dst = np.zeros_like(img_src)
img_dst2 = img_dst.copy()
img_tmp = cv2.pyrDown(img_src)
h,w = img_tmp.shape[:2]
img_dst[:h,:w,:] = img_tmp
img_dst = cv2.hconcat([img_src,img_dst])

img_tmp2 = cv2.pyrDown(img_tmp)
h,w = img_tmp2.shape[:2]
img_dst2[:h,:w,:] = img_tmp2
img_dst = cv2.hconcat([img_dst,img_dst2])

cv2.imshow('src',img_dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
# video 1/2크기로 변환후 hconcat
capture = cv2.VideoCapture('IVE.mp4')
if capture.isOpened() == False:
    print("동영상을 열수 없음")
    exit(1)


while cv2.waitKey(33) < 0: #아무키나 누르면 종료
    ret, img_src = capture.read()
    img_src = cv2.pyrDown(img_src)
    img_dst = np.zeros_like(img_src)
    img_tmp = cv2.pyrDown(img_src)
    h,w = img_tmp.shape[:2]
    img_dst[h//2:3*h//2,w//2:3*w//2,:] = img_tmp
    img_dst = cv2.hconcat([img_src,img_dst])
    # 동영상 끝까지 읽으면 처음으로 돌아감
    if capture.get(cv2.CAP_PROP_POS_FRAMES) == capture.get(cv2.CAP_PROP_FRAME_COUNT):
        capture.set(cv2.CAP_PROP_POS_FRAMES, 0)
    cv2.imshow('dst', img_dst)
capture.release()
cv2.destroyAllWindows()
```
